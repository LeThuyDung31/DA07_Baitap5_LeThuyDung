```python
import pandas as pd
import numpy as np
table_housing = pd.read_csv("VN_housing_dataset.csv")
print (table_housing.shape)
table_housing_2 = table_housing.drop(['Ngày','Địa chỉ','Quận','Huyện','Loại hình nhà ở','Giấy tờ pháp lý','Số tầng','Số phòng ngủ'],axis='columns')
table_housing_2.shape
print(table_housing_2.head())
```

```python
table_housing_2.isnull().sum()
```

```python
table_clean= table_housing_2.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
price_m2 = table_clean['Giá/m2']
new_price_m2 = []

for item in price_m2:
    new_price_m2.append(item)
    if(item.find("tỷ") != -1):
        item_split = item.split()
        new_item = str("{:,}".format((int(item_split[0].replace(',','.').replace('.',''))) * 1000)) + " triệu/m²"
        print(new_item)
        new_price_m2.append(new_item)
```

```python
import datetime as dt
from datetime import datetime
dates = table_housing ['Ngày']
date_convert = []
for i in dates:
    date_config = dt.datetime.strptime(str(i),"%Y-%m-%d")
    new_date = date_config.strftime("%d-%m-%Y")
    date_convert.append(new_date)
date_convert
```
